```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def plot_rosenbrock(a,b):
    x = np.arange(0.1,1.1,0.01)
    y = np.arange(0.1,1.1,0.01)
    x_, y_ = np.meshgrid(x,y,indexing='ij')
    z_ = -((a-x_)**2+b*(y_-x_**2)**2)  # 画图所要表现出来的主函数
    fig = plt.figure(figsize=(10, 10), facecolor='white') #创建图片
    sub = fig.add_subplot(111, projection='3d')# 添加子图，
    surf = sub.plot_surface(x_, y_, z_, cmap=plt.cm.brg) #绘制曲面,cmap=plt.cm.brg并设置颜色cmap
    cb = fig.colorbar(surf, shrink=0.8, aspect=15) #设置颜色棒

    sub.set_xlabel(r"x1 axis")
    sub.set_ylabel(r"x2 axis")
    sub.set_zlabel(r"z axis")
    plt.savefig('./funcs/rosenbrock.png')
```

```python
a=1
b=100
plot_rosenbrock(1,100)
```

```python
def plot_rastrigin(a,scale_factor=1):
    x = np.arange(-0.1,0.9,0.01)
    y = np.arange(-0.1,0.9,0.01)
    x_, y_ = np.meshgrid(x,y,indexing='ij')
    z_ = -(a*2 + (scale_factor*x_)**2 + (scale_factor*y_)**2 - a*np.cos(2*np.pi*scale_factor*x_) - a*np.cos(2*np.pi*scale_factor*y_))  # 画图所要表现出来的主函数
    fig = plt.figure(figsize=(10, 10), facecolor='white') #创建图片
    sub = fig.add_subplot(111, projection='3d')# 添加子图，
    surf = sub.plot_surface(x_, y_, z_, cmap=plt.cm.brg) #绘制曲面,cmap=plt.cm.brg并设置颜色cmap
    cb = fig.colorbar(surf, shrink=0.8, aspect=15) #设置颜色棒

    sub.set_xlabel(r"x1 axis")
    sub.set_ylabel(r"x2 axis")
    sub.set_zlabel(r"z axis")
    plt.savefig('./funcs/rastrigin.png')
```

```python
a=10
plot_rastrigin(10,scale_factor=2)
```

```python
def plot_ackley(a,b,scale_factor):
    x = np.arange(-0.1,0.9,0.01)
    y = np.arange(-0.1,0.9,0.01)
    x_, y_ = np.meshgrid(x,y,indexing='ij')
    z_ = -(-a * np.exp(-b * np.sqrt(1.0/2 * ((scale_factor*x_)**2+(scale_factor*y_)**2))) - np.exp(1.0/2 * (np.cos(2 * np.pi * scale_factor * x_)+np.cos(2 * np.pi * scale_factor * y_))) + a + np.e)  # 画图所要表现出来的主函数
    fig = plt.figure(figsize=(10, 10), facecolor='white') #创建图片
    sub = fig.add_subplot(111, projection='3d')# 添加子图，
    surf = sub.plot_surface(x_, y_, z_, cmap=plt.cm.brg) #绘制曲面,cmap=plt.cm.brg并设置颜色cmap
    cb = fig.colorbar(surf, shrink=0.8, aspect=15) #设置颜色棒

    sub.set_xlabel(r"x1 axis")
    sub.set_ylabel(r"x2 axis")
    sub.set_zlabel(r"z axis")
    plt.savefig('./funcs/ackley.png')
```

```python
a=20
b=0.2
plot_ackley(a,b,scale_factor=2)
```
