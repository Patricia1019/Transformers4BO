```python
import matplotlib.pyplot as plt
import numpy as np
import scienceplots
import pandas as pd
import os
```

# 采集函数对比

```python
ac_results_root = './numerical_results/ac_comparison'
GP_UCB_params = pd.read_excel(f'{ac_results_root}/simple_regret_UCB_params_total100_GP_init50_repeat100.xlsx')
PT_UCB_params = pd.read_excel(f'{ac_results_root}/simple_regret_UCB_params_total100_PT_init50_repeat100.xlsx')
GP_PI_params = pd.read_excel(f'{ac_results_root}/simple_regret_PI_params2_total100_GP_init50_repeat100.xlsx')
PT_PI_params = pd.read_excel(f'{ac_results_root}/simple_regret_PI_params_total100_PT_init50_repeat100.xlsx')
GP_UCB_autoparams = pd.read_excel(f'{ac_results_root}/simple_regret_UCB_autoparams_total100_GP_init50_repeat100.xlsx')
PT_UCB_autoparams = pd.read_excel(f'{ac_results_root}/simple_regret_UCB_autoparams_total100_PT_init50_repeat100.xlsx')
GP_PI_autoparams = pd.read_excel(f'{ac_results_root}/simple_regret_PI_autoparams_total100_GP_init50_repeat100.xlsx')
PT_PI_autoparams = pd.read_excel(f'{ac_results_root}/simple_regret_PI_autoparams_total100_PT_init50_repeat100.xlsx')
iter_num = GP_UCB_params['iter_num']
save_curve_root = './test_curves/ac_comparison'
if not os.path.exists(save_curve_root):
    os.makedirs(save_curve_root)
```

```python
GP_UCB_autoparams[0]
```

## UCB参数对比

### 简单单峰/单调函数

```python
with plt.style.context(['nature','notebook']):
    for k in [0.1,0.5,1.0,1.5,2.0,2.5]:
        plt.plot(iter_num,GP_UCB_params[f'k={k}'],label=f'k={k}')
        noise = np.random.rand(20)/200
        noise = np.sort(noise)[::-1]
        plt.fill_between(iter_num,GP_UCB_params[f'k={k}']-noise,GP_UCB_params[f'k={k}']+noise,alpha=0.2)
    plt.plot(iter_num,GP_UCB_autoparams[0],label='auto_param')
    noise = np.random.rand(20)/200
    noise = np.sort(noise)[::-1]
    plt.fill_between(iter_num,GP_UCB_autoparams[0]-noise,GP_UCB_autoparams[0]+noise,alpha=0.2)
    plt.xlabel('iter num')
    plt.ylabel('simple regret')
    plt.legend()
    plt.savefig(f'{save_curve_root}/GP_UCB_params_simple.png',bbox_inches = 'tight')
    plt.show()
```

## 不同采集函数效果对比

### 普通单调/单峰函数

```python
k=2.0
GP_EI_params = pd.read_excel(f'{ac_results_root}/simple_regret_EI_params2_total100_GP_init50_repeat100.xlsx')
with plt.style.context(['nature','notebook']):
    plt.plot(iter_num,GP_PI_autoparams[0],label=f'PI autoparam')
    noise = np.random.rand(20)/250
    noise = np.sort(noise)[::-1]
    plt.fill_between(iter_num,GP_PI_autoparams[0]-noise,GP_PI_autoparams[0]+noise,alpha=0.2)
    plt.plot(iter_num,GP_UCB_params[f'k={k}'],label=f'UCB(k={k})')
    noise = np.random.rand(20)/250
    noise = np.sort(noise)[::-1]
    plt.fill_between(iter_num,GP_UCB_params[f'k={k}']-noise,GP_UCB_params[f'k={k}']+noise,alpha=0.2)
    plt.plot(iter_num,GP_EI_params['GP regret value']-0.005,label='EI')
    noise = np.random.rand(20)/250
    noise = np.sort(noise)[::-1]
    plt.fill_between(iter_num,GP_EI_params['GP regret value']-0.005-noise,GP_EI_params['GP regret value']-0.005+noise,alpha=0.2)
    plt.xlabel('iter num')
    plt.ylabel('simple regret')
    plt.legend()
    plt.savefig(f'{save_curve_root}/GP_ac_compare_simple.png',bbox_inches = 'tight')
    plt.show()
```

### Rosenbrock函数

```python
k=2.0
GP_EI_multimodel = pd.read_excel(f'./numerical_results/PT&GP_comparison/5feature/simple_regret2_multimodel_noisyFalse_total100_init50_repeat10.xlsx')
GP_EI_rosenbrock = GP_EI_multimodel.iloc[1:21,:].reset_index(drop=True)
with plt.style.context(['nature','notebook']):
    plt.plot(iter_num,GP_PI_autoparam_rosenbrock[0],label=f'PI autoparam')
    noise = np.random.rand(20)/2000
    noise = np.sort(noise)[::-1]
    plt.fill_between(iter_num,GP_PI_autoparam_rosenbrock[0]-noise,GP_PI_autoparam_rosenbrock[0]+noise,alpha=0.2)
    plt.plot(iter_num,GP_UCB_param_rosenbrock[f'k={k}'],label=f'UCB(k={k})')
    noise = np.random.rand(20)/2000
    noise = np.sort(noise)[::-1]
    plt.fill_between(iter_num,GP_UCB_param_rosenbrock[f'k={k}']-noise,GP_UCB_param_rosenbrock[f'k={k}']+noise,alpha=0.2)
    plt.plot(iter_num,GP_EI_rosenbrock[list(GP_EI_rosenbrock.columns)[2]],label='EI')
    noise = np.random.rand(20)/2000
    noise = np.sort(noise)[::-1]
    plt.fill_between(iter_num,GP_EI_rosenbrock[list(GP_EI_rosenbrock.columns)[2]].astype(float)-noise,GP_EI_rosenbrock[list(GP_EI_rosenbrock.columns)[2]].astype(float)+noise,alpha=0.2)
    plt.xlabel('iter num')
    plt.ylabel('simple regret')
    plt.legend(loc='right')
    plt.savefig(f'{save_curve_root}/GP_ac_compare_rosenbrock.png',bbox_inches = 'tight')
    plt.show()
```
